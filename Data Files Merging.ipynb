```python
import pandas as pd
```

```python
places_cleaned = pd.read_csv('places_cleaned.csv')
place_recording_binary = pd.read_csv('place_recording_binary.csv')
place_work_binary = pd.read_csv('place_work_binary.csv')
```

```python
merged_df = places_cleaned.merge(place_recording_binary, left_on='id', right_on='place_id', how='left')
merged_df = merged_df.merge(place_work_binary, left_on='id', right_on='place_id', how='left')
```

```python
merged_df
```

```python
merged_df.drop(columns = ['place_id_x', 'place_id_y'], inplace=True)
```

```python
merged_df.columns
```

```python
merged_df = merged_df.drop(columns=[col for col in merged_df.columns if "Unnamed: 0" in col], errors='ignore')
```

```python
merged_df
```

```python
merged_df = merged_df.rename(columns={"arranged at_x": "arranged at_recording", "arranged at_y": "arranged at_work"})
```
